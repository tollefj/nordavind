```python
from datasets import load_dataset

mosaic = load_dataset("tollefj/nob-mosaikk-instruct")
nor = load_dataset("tollefj/nor-instruct")
```

```python
# combine "instruction" and "input" into a space-separated string
def combine_instruction_and_input(example):
    return {"instruction": example["instruction"] + " " + example["input"]}

nor = nor.map(combine_instruction_and_input)
```

```python
# drop "input" column
nor = nor.remove_columns("input")
nor = nor.rename_column("output", "response")
```

```python
# concat
from datasets import concatenate_datasets, DatasetDict

combined = DatasetDict({
    "train": concatenate_datasets([mosaic["train"], nor["train"]]),
    "test": concatenate_datasets([mosaic["test"], nor["test"]]),
})
```
